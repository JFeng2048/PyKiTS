1. 异步函数

````python
async def magic_demo():
    print("🎪 开始魔术表演")

    # 魔术1：创建3个并发任务
    task_a = asyncio.create_task(asyncio.sleep(3, "A完成"))
    task_b = asyncio.create_task(asyncio.sleep(2, "B完成"))
    task_c = asyncio.create_task(asyncio.sleep(1, "C完成"))

    print("📦 所有任务都已创建，开始等待...")

    # 魔术2：await 的调度过程
    for i, task in enumerate([task_a, task_b, task_c], 1):
        result = await task  # 🔄 关键点：这里会先完成最快的任务！
        print(f"第{i}个await完成: {result}")

    print("🎉 所有魔术完成！")
asyncio.run(magic_demo())
````

```
任务并发执行：当你使用 asyncio.create_task() 创建任务时，这些任务会并发执行。
await 的行为：虽然任务是并发执行的，但 await 语句是按顺序执行的。在你的代码中，await 是按 [task_a, task_b, task_c] 的顺序进行的。
关键点：即使 task_c 最先完成，但因为你按顺序 await task_a、await task_b、await task_c，所以结果会按照你 await 的顺序返回，而不是任务完成的顺序。
```

````python
async def magic_demo():
    print("🎪 开始魔术表演")

    # 魔术1：创建3个并发任务
    task_a = asyncio.create_task(asyncio.sleep(3, "A完成"))
    task_b = asyncio.create_task(asyncio.sleep(2, "B完成"))
    task_c = asyncio.create_task(asyncio.sleep(1, "C完成"))

    print("📦 所有任务都已创建，开始等待...")

    # 使用 as_completed 按完成顺序处理任务
    tasks = [task_a, task_b, task_c]
    for i, task in enumerate(asyncio.as_completed(tasks), 1):
        result = await task
        print(f"第{i}个await完成: {result}")

    print("🎉 所有魔术完成！")
asyncio.run(magic_demo())
````

✅ asyncio.create_task() = 立即开始执行（并发）

✅ await task = 等待这个特定任务完成

✅ 任务在后台并发执行，但await按代码顺序等待

````python
import asyncio
import time

async def tracked_task(name, seconds):
    start = time.time()
    print(f"🕒 [{time.time()-start:.1f}s] 任务{name}开始执行")
    await asyncio.sleep(seconds)
    print(f"✅ [{time.time()-start:.1f}s] 任务{name}完成")
    return f"{name}完成"

async def magic_demo_clear():
    start_time = time.time()
    print(f"🚀 [{time.time()-start_time:.1f}s] 开始魔术表演")

    # 创建任务 - 立即开始执行！
    task_a = asyncio.create_task(tracked_task("A", 3))
    task_b = asyncio.create_task(tracked_task("B", 2))
    task_c = asyncio.create_task(tracked_task("C", 1))

    print(f"📦 [{time.time()-start_time:.1f}s] 所有任务已创建")

    # 按顺序等待，但任务已经在并发执行
    print(f"⏳ [{time.time()-start_time:.1f}s] 开始等待任务A...")
    result_a = await task_a
    print(f"📥 [{time.time()-start_time:.1f}s] 收到: {result_a}")

    print(f"⏳ [{time.time()-start_time:.1f}s] 开始等待任务B...")
    result_b = await task_b
    print(f"📥 [{time.time()-start_time:.1f}s] 收到: {result_b}")

    print(f"⏳ [{time.time()-start_time:.1f}s] 开始等待任务C...")
    result_c = await task_c
    print(f"📥 [{time.time()-start_time:.1f}s] 收到: {result_c}")

    print(f"🎉 [{time.time()-start_time:.1f}s] 所有完成!")

await magic_demo_clear()

# 运行结果：
# 🚀 [0.0s] 开始魔术表演
# 🕒 [0.0s] 任务A开始执行
# 🕒 [0.0s] 任务B开始执行
# 🕒 [0.0s] 任务C开始执行
# 📦 [0.0s] 所有任务已创建
# ⏳ [0.0s] 开始等待任务A...
# ✅ [1.0s] 任务C完成
# ✅ [2.0s] 任务B完成
# ✅ [3.0s] 任务A完成
# 📥 [3.0s] 收到: A完成
# ⏳ [3.0s] 开始等待任务B...
# 📥 [3.0s] 收到: B完成  # 立即返回，因为B早就完成了
# ⏳ [3.0s] 开始等待任务C...
# 📥 [3.0s] 收到: C完成  # 立即返回，因为C早就完成了
# 🎉 [3.0s] 所有完成!
````
